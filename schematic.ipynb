```python
import schemdraw
import schemdraw.elements as elm
with schemdraw.Drawing(file='schematic.svg') as d:
    elm.Resistor().label('100KΩ')
    elm.Capacitor().down().label('0.1μF', loc='bottom')
    elm.Line().left()
    elm.Ground()
    elm.SourceV().up().label('10V')
```

```python
from schemdraw import dsp
with schemdraw.Drawing() as d:
    d.config(fontsize=12)
    dsp.Antenna()
    dsp.Line().right(d.unit/4)#
    dsp.Amp().fill('lightblue').label('LNA')#
    dsp.Line().length(d.unit/3)
    mix = dsp.Mixer().fill('navajowhite').label('Mixer')
    dsp.Line().at(mix.S).down(d.unit/3)
    dsp.Oscillator().right().anchor('N').fill('navajowhite').label('Local\nOscillator', 'right', ofst=.2)
    dsp.Line().at(mix.E).right(d.unit/3)
    dsp.Amp().fill('lightblue').label('Power\namplifier')
    dsp.Line().length(d.unit/4)
    dsp.Antenna()
```

```python
from schemdraw import dsp
with schemdraw.Drawing() as d:
    d.config(fontsize=12)
    dsp.Antenna()
    dsp.Line().right(d.unit/4)#
    dsp.Amp().fill('lightblue').label('LNA')#
    dsp.Line().length(d.unit/3)
    mix = dsp.Mixer().fill('navajowhite').label('D/C')
    dsp.Line().at(mix.S).down(d.unit/3)
    dsp.Oscillator().right().anchor('N').fill('navajowhite').label('Local\nOscillator', 'right', ofst=.2)
    dsp.Line().at(mix.E).right(d.unit/3)
    dsp.Amp().fill('lightblue').label('IF\namplifier')
    dsp.Line().length(d.unit/3)
    mix = dsp.Mixer().fill('navajowhite').label('U/C')
    dsp.Line().at(mix.S).down(d.unit/3)
    dsp.Oscillator().right().anchor('N').fill('navajowhite').label('Local\nOscillator', 'right', ofst=.2)
    dsp.Line().at(mix.E).right(d.unit/3)
    dsp.Amp().fill('lightblue').label('HPA')#
    dsp.Line().length(d.unit/3)

    dsp.Antenna()
```
