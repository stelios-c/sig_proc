```python
from scipy import signal
import matplotlib.pyplot as plt
```

# Moving Average

```python
ma3=signal.TransferFunction([1/4, 1/4, 1/4,1/4],1,  dt=1)
```

```python
ma3
```

```python
ma3.to_zpk()
```

```python
w, H = signal.dfreqresp(ma3)
```

```python
plt.figure()
plt.plot(H.real, H.imag, "b")
plt.plot(H.real, -H.imag, "r")
plt.show()
```

```python
w, mag, phase = signal.dbode(ma3)
```

```python
plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase)  # Bode phase plot
plt.show()
```
